```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# To load the cleaned dataset
df = pd.read_excel(r"C:\Users\Riri\Downloads\oppurtunity_data_cleaned.xlsx")
df
```

```python
# Lets et aesthetic style for the charts using seaborn
sns.set_theme(style="whitegrid")
```

```python
# Convert to datetime that is the dates to specific months for easier visualixation
df['created_at'] = pd.to_datetime(df['created_at'])

# Extract Year-Month
df['created_month'] = df['created_at'].dt.to_period('M')

# Group and plot
monthly_trends = df.groupby('created_month').size()
monthly_trends.plot(kind='line', figsize=(10, 5), title='Opportunity Sign-up Trends Over Time')
plt.ylabel('Number of Opportunities')
plt.show()
```

```python
# Visualize top locations (excluding 'Unknown' if necessary)
top_locations = df['location'].value_counts().head(10)
top_locations.plot(kind='bar', figsize=(10, 5), title='Top 10 Opportunity Locations')
plt.ylabel('Count')
plt.show()
```

```python
category_counts = df['category'].value_counts()
category_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), title='Distribution by Category')
plt.show()
```

```python
# Create a histogram to view the distribution
sns.histplot(df['micro_scholarship'], bins=30, kde=True)
plt.title('Distribution of Micro-scholarship Amounts')
plt.show()

# Get summary statistics
print(df['micro_scholarship'].describe())
```
