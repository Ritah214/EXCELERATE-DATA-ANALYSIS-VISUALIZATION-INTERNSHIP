```python
#Now we have need to fill the cells with missing data with a placeholder such as 'None' or 'Null' to retain the rows with missing values
#Remember we now have columns with valid values more than 30%
#The cells that we can fill are the ones that we really don't know examples someone's location and their category we don't know if itts carrer or competition etc 
df1_clean['location'] = df1_clean['location'].fillna('Unknown')
df1_clean['category'] = df1_clean['category'].fillna('Unknown')
```

```python
#The warning is normal we need to let Pandas where to look since the df1_clean is already a data frame and we are trying to edit that same df 
#Pandas recommends using .loc
df1_clean.loc[df1_clean['location'].isna(), 'location'] = 'Unknown'
df1_clean.loc[df1_clean['category'].isna(), 'category'] = 'Unknown'
df1_clean.loc[df1_clean['currency_type'].isna(), 'currency_type'] = 'Unknown'
df1_clean.loc[df1_clean['duration_type'].isna(), 'duration_type'] = 'Unknown'
df1_clean.loc[df1_clean['role'].isna(), 'role'] = 'Unknown'
df1_clean.loc[df1_clean['name'].isna(), 'name'] = 'Not Provided' 
df1_clean.loc[df1_clean['short_description'].isna(), 'short_description'] = 'Not Provided'
df1_clean.loc[df1_clean['long_description'].isna(), 'long_description'] = 'Not Provided'
df1_clean.loc[df1_clean['role_responsibility'].isna(), 'role_responsibility'] = 'Not Provided'

# 1. Fill missing ID-based columns with 'None'
json_cols = ['Badge', 'CareerAddOn', 'Cohort', 'Eligibility', 'Reward', 'Testimonial', 'Panellist', 'tracking_questions']
for col in json_cols:
    if col in df1_clean.columns:
        df1_clean[col] = df1_clean[col].fillna('None')

# 2. Fill numerical columns with 0
df1_clean['fee'] = pd.to_numeric(df1_clean['fee'], errors='coerce').fillna(0)
df1_clean['microscholarship'] = df1_clean['microscholarship'].fillna(0)
df1_clean['duration'] = df1_clean['duration'].fillna(0)

# 3. Fill missing categorical/link columns
df1_clean['code'] = df1_clean['code'].fillna('Unknown')
df1_clean['image_link'] = df1_clean['image_link'].fillna('No Image Provided')
df1_clean['is_auto_approve'] = df1_clean['is_auto_approve'].replace({'NAN': False, 'None': False})
df1_clean['is_auto_approve'] = df1_clean['is_auto_approve'].fillna(False).astype(bool)
date_cols = ['created_at', 'modified_at', 'last_date_to_apply']
df1_clean[date_cols] = df1_clean[date_cols].ffill()
print("Missing values after full cleanup:")
print(df1_clean.isnull().sum())
```

```python
#We also need to change the date string which are scientific notations that represent milliseconds (1.66E+12) to date format
def parse_timestamp(val):
    try:
        return pd.to_datetime(float(val), unit='ms', errors='coerce')
    except:
        return pd.NaT
```

```python
#Now we need to apply these function to our date columns ie created_at, modified_at
df1_clean = df1_clean.copy()
#This is to ensure the .loc warning doestn't appear
def parse_timestamp(val):
    try:
        # Convert the string '1.66E+12' into a number
        num = float(val)
        # Convert the number into a datetime object
        return pd.to_datetime(num, unit='ms')
    except (ValueError, TypeError):
        # If the cell is completely empty or corrupted, return NaT safely
        return pd.NaT

# Now your original code will work perfectly
df1_clean['created_at'] = df1_clean['created_at'].apply(parse_timestamp)
df1_clean['modified_at'] = df1_clean['modified_at'].apply(parse_timestamp)
df1_clean['last_date_to_apply'] = df1_clean['last_date_to_apply'].apply(parse_timestamp)
```

```python
#To address the True andfalse issues, years we need to enforce standard casing and format
df1_clean['category'] = df1_clean['category'].astype(str).str.title()
df1_clean['duration_type'] = df1_clean['duration_type'].astype(str).str.lower()
df1_clean['is_auto_approve'] = df1_clean['is_auto_approve'].astype(str).str.upper()
```

```python
 #To ensure consistency we need to ensure that the once that have not been changed are changed effectively TRUE will change to True
df1_clean['is_auto_approve'] = df1_clean['is_auto_approve'].replace({'TRUE': True, 'FALSE': False})
```

```python
#We need to make these {"sk":"Badge#0000000010267T036230A6RZB4","crea... and <html><body><ul><li>Leadin... readable
#We need to import REGULAR EXPRESSIONS AS RE
# It is a built-in Python library used for searching, matching, and manipulating specific patterns of text.
#We will use it to Stripping HTML tags and Extracting IDs which are in JSON
import re
```

```python
#Starting with html
def remove_html_tags(text):
    if isinstance(text, str):
        clean = re.compile('<.*?>')
        return re.sub(clean, '', text)
    return text

# Apply the function to relevant columns
df1_clean['role_responsibility'] = df1_clean['role_responsibility'].apply(remove_html_tags)
df1_clean['long_description'] = df1_clean['long_description'].apply(remove_html_tags)
```

```python
#Next the JSON
def extract_sk_ids(text):
    if pd.isna(text):
        return 'None'
    # This regex finds all text placed immediately after "sk":"
    matches = re.findall(r'"sk":"([^"]+)"', str(text))
    
    if matches:
        return ', '.join(matches) # Joins multiple IDs with a comma if there are more than one
    return 'None'

# 2. We define the columns that have this messy JSON format
json_columns = [
    'Badge', 'Cohort', 'Eligibility', 'Testimonial', 
    'Reward', 'CareerAddOn', 'Panellist', 'tracking_questions'
]

# 3. Apply the function to clean those columns in your dataframe
for col in json_columns:
    if col in df1_clean.columns:
        df1_clean[col] = df1_clean[col].apply(extract_sk_ids)
```
